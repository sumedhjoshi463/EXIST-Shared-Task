### Splitting the data 

```python
def split_training_data(df):
    split_data_samples = []
    for index, data_sample in df.iterrows():
    # Iterate over each annotator and create a new data sample
        for i in range(data_sample["number_annotators"]):
            split_data = {}
            split_data["id_EXIST"] = data_sample["id_EXIST"]
            split_data["lang"] = data_sample["lang"]
            split_data["tweet"] = data_sample["tweet"]
            split_data["annotator"] = data_sample["annotators"][i]
            split_data["gender_annotator"] = data_sample["gender_annotators"][i]
            split_data["age_annotator"] = data_sample["age_annotators"][i]
            split_data["ethnicity_annotator"] = data_sample["ethnicities_annotators"][i]
            split_data["study_level_annotator"] = data_sample["study_levels_annotators"][i]
            split_data["country_annotator"] = data_sample["countries_annotators"][i]
            split_data["label_task1"] = data_sample["labels_task1"][i]
            split_data["label_task2"] = data_sample["labels_task2"][i]
            split_data["label_task3"] = data_sample["labels_task3"][i]
            split_data["split"] = data_sample["split"]

            # Append the split data sample to the list
            split_data_samples.append(split_data)
    
    new_df = pd.DataFrame(split_data_samples)
    # Print the split data samples
    return new_df
```
