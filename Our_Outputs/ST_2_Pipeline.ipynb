```python
from transformers import pipeline


class Pipeline:
    def __init__(self):
        self.model1 = pipeline("text-classification", model="annahaz/xlm-roberta-base-misogyny-sexism-indomain-mix-bal")
        self.model2 = pipeline("text-classification", model="SumedhJoshi/idk")
        self.model3 = pipeline("text-classification", model="Abhi964/st_sexism_task_3")

    def get_result(self, id, text):
        try:
            answer = self.model1(text)

            if answer[0]['label'] == '1':
                answer2 = self.model2(text)
                answer3 = self.model3(text)
                return id, answer[0]['label'], answer2[0]['label'], answer3[0]['label']
            else:
                return id, answer[0]['label'], '-', '-'

        except Exception as e:
            raise e
```

```python
obj = Pipeline()
```

```python
res = obj.get_result(2003,"You look like a whore. Stop projecting!")
```

```python
test_data = df[['id_EXIST','English']]
```

```python
type(test_data)
```

```python
from tqdm import tqdm

task1f =[]
task2f =[]
task3f =[]

obj = Pipeline()

for index, row in tqdm(test_data.iterrows(), total=len(test_data)):
    # Call your function with id and text
    answer = obj.get_result(row['id_EXIST'], row['English'])

    task1f.append({"id":answer[0], "value":answer[1], "test_case":"EXIST2024"})
    task2f.append({"id":answer[0], "value":answer[2], "test_case":"EXIST2024"})
    task3f.append({"id":answer[0], "value":[answer[3]], "test_case":"EXIST2024"})
```
